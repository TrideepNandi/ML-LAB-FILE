The code df1 = df.fillna(value=values) in pandas replaces all null or missing values in the DataFrame df with the corresponding values specified in the values dictionary, and assigns the resulting DataFrame to df1.

```python
values={'Price': '$20','Discount': 'some values','Rating': '3 star','Currency' : 'INR', 'Features' : 'short'}
df1 = df.fillna(value=values)
df1
```

The ffill1= df.fillna(method = "ffill") line of code in pandas fills the null or missing values in the DataFrame df with the value of the previous row (forward fill). The resulting DataFrame is then assigned to ffill1. This is particularly useful in time series data where you might want to fill missing values with the last known value.

```python
ffill1= df.fillna(method = "ffill")
ffill1
```

The bfill1= df.fillna(method = "bfill") line of code in pandas fills the null or missing values in the DataFrame df with the value of the next row (backward fill). The resulting DataFrame is then assigned to bfill1. This is particularly useful in time series data where you might want to fill missing values with the next known value.

```python
bfill1= df.fillna(method = "bfill")
bfill1
```

The code df1 = df.fillna(value=values, limit = 2) in pandas replaces the first two null or missing values in each column of the DataFrame df with the corresponding values specified in the values dictionary. The resulting DataFrame is then assigned to df1.  

The limit=2 parameter is used to restrict the operation to the first two missing values in each column. This means that if a column has more than two missing values, only the first two will be replaced, and the rest will remain as missing values. This can be useful when you want to fill in missing values but maintain some level of missing data for analysis.

```python
values={'Price': '$20','Discount': 'some values','Rating': '3 star','Currency' : 'INR', 'Features' : 'short'}
df1 = df.fillna(value=values, limit = 2)
df1
```
